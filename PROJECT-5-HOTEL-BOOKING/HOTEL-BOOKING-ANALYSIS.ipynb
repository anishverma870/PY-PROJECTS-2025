# **DATA ANALYSIS AND VISUALISATION**

```python
cancel_reservation = hb['is_canceled'].value_counts(normalize=True)*100 
print(cancel_reservation)

```

```python
plt.figure(figsize=(5,3))
sns.countplot(data =hb,x= 'is_canceled',palette='cool')
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data= hb,x = 'hotel',hue='is_canceled',palette='cool')
plt.show()
```

```python
resort_hotel = hb[hb['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
city_hotel = hb[hb['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
hb['months'] = hb['reservation_status_date'].dt.month
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='months',hue='is_canceled',data=hb,palette='cool')
plt.legend(['not_legend','cancelled'])
plt.show()
```

```python
adr_data = hb[hb['is_canceled']==1].groupby('months')['adr'].sum().reset_index()
```

```python
sns.barplot(x='months',y='adr',data=adr_data,palette='cool')
plt.show()
```

```python
cancelled = hb[hb['is_canceled']==1]
```

```python
top_10_country = cancelled['country'].value_counts()[:10]
plt.pie(x=top_10_country,labels=top_10_country.index,autopct='%.2f')
plt.title('top 10 country with reservation cancelled ',color = 'g')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=cancelled , x=cancelled['is_repeated_guest'],palette='cool')
plt.show()
```

```python
hb['market_segment'].value_counts(normalize=True)*100
```

```python
cancelled['market_segment'].value_counts(normalize=True)*100
sns.countplot(data=cancelled,x=cancelled['market_segment'],palette='cool')
plt.show()
```
