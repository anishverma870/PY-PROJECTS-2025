### SEPERATION OF MONTH AND YEARS

```python
st['Order Date'] = pd.to_datetime(st['Order Date'])
st['Order_Month'] = st['Order Date'].dt.month
st['Order_Year'] = st['Order Date'].dt.year
```

## **UNIVARIATE ANALYSIS**

```python
sns.countplot(x='Order_Month', data=st)
plt.title('Orders by Month')
plt.show()
```

Which segment is dominant?

```python
sns.countplot(x= st['Segment'],palette='cool')
plt.show()
```

Most/least used delivery methods?

```python
sns.countplot(x=st['Ship Mode'],palette='cool')
plt.title('Most/least used delivery methods',color='red')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x=st['State'],palette='cool')
plt.title('ORDER BY STATE',color='blue')
plt.xticks(rotation = 90)
plt.show()
```

```python
data = st.groupby('Order Date')['Profit'].sum()
```

```python
plt.figure(figsize=(20,6))
sns.lineplot(data=data)
plt.show()
```

```python
import numpy as np
st['Log_Sales'] = np.log1p(st['Sales'])  # log1p = log(1 + x)

sns.histplot(st['Log_Sales'], kde=True)
plt.title('Log-Transformed Sales Distribution')
plt.show()
```
